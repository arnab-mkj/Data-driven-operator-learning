### Universal Approximation Theorem

The **Universal Approximation Theorem** is a fundamental result in neural network theory. It states that a feedforward neural network with at least one hidden layer containing a finite number of neurons, and with appropriate activation functions (such as sigmoid, ReLU, or tanh), can approximate any continuous function on a compact subset of ${R}^n $ to any desired degree of accuracy, given enough neurons in the hidden layer.

#### Key Points

- The theorem applies to **feedforward neural networks** (also called multilayer perceptrons).
- The network must have at least **one hidden layer**.
- The activation function must be **non-constant, bounded, and continuous** (sigmoid and tanh are common examples; later results also include ReLU).
- The theorem guarantees the **existence** of such a network, but does not say how to find the right parameters or how many neurons are needed in practice.

#### Mathematical Statement

Let \\( f: {R}^n --> {R} \\) be a continuous function on a compact subset $ K \subset {R}^n $. For any $ \epsilon > 0 $, there exists a neural network \( F \) with one hidden layer such that

$$
|f(x) - F(x)| < (\epsilon) \quad \text{for all } x \in K
$$

The universal approximation theorem states that neural networks can be used to approximate any continuous
 function to arbitrary accuracy if no constraint is placed on the width and depth of the hidden layers

# Universal Approximation Theorem for Operator

1. **The Core Idea**: The theorem states that a neural network with a specific structure can accurately approximate any linear or nonlinear continuous operator. An operator $G$ is a mathematical mapping that takes an entire function $u$ as input and outputs another function $G(u)$.

2. **What is Approximated**: The theorem focuses on approximating the value of the output function $G(u)$ at a specific location $y$, denoted as $G(u)(y)$. This value $G(u)(y)$ is a real number.

3. **The Inputs to the Approximating Network**: To approximate $G(u)(y)$, the network needs two types of information as input:
* Information about the input function $u$. Since a network cannot directly process an entire continuous function, the theorem's structure (and the DeepONet architecture it inspires) uses the values of the input function $u$ evaluated at a fixed number of specific locations, called "sensors" {x1, x2, ..., xm} . These m sensor locations {x1, ..., xm} are the same for all input functions $u$ considered.
* The location $y$ where the output value $G(u)(y)$ is desired. This location $y$ can be a vector in ${R^d}$.

4. **The Network Structure**: describes a specific architecture (given by Eq. 1) that can achieve this approximation. This structure has two main parts that are combined:
* A "*branch*" part that processes the sensor values of the input function. This part takes the vector of values [u(x1), u(x2), ..., u(xm)] as input.
* A "*trunk*" part that processes the output location y. This part takes the vector y as input.
* Both parts utilize an activation function σ. The theorem requires $σ$ to be a continuous non-polynomial function. Many common activation functions used in neural networks, like sigmoid, tanh, and ReLU, satisfy this condition.
* The overall output G(u)(y) is approximated by combining the outputs of the branch and trunk parts, typically through summation or multiplication of corresponding components. The formula in Theorem 1 (Eq. 1) shows a sum of terms, where each term is a product of an output from the branch processing $u$ $(x_j)$ and an output from the trunk processing $y$.

5. **The Guarantee**: The theorem guarantees that for any desired level of accuracy (represented by ε > 0), you can find a network of this structure (with a sufficient number of neurons, represented by integers n, p, m, and specific weights and biases) such that the network's output is arbitrarily close (within ε) to the true value $G(u)(y)$ for all possible input functions $u$ within a specified set $V$ and all possible output locations $y$ within a specified set $K2$.

6. **Conditions for the Theorem to Hold**: The theorem holds under the assumption that:
* $σ$ is a continuous non-polynomial function.
* The space where input functions are defined (X) is a Banach space.
* The sets $K1$ (where $u$ is defined and sensors are located) and $K2$ (where $y$ is located) are compact.
* The set of input functions $V$ considered is a compact set in the space of continuous functions $C(K1)$.
* The operator $G$ is nonlinear and continuous

**Parameter definitions:**

- $G(u)(y)$: The target function evaluated at $y$ with input $u$.
- $p$: Number of terms in the outer sum (number of "trunks").
- $n$: Number of terms in the inner sum (number of "branches" per trunk).
- $c_i^k$: Coefficient for the $i$-th branch in the $k$-th trunk.
- $\sigma(\cdot)$: Activation function (e.g., sigmoid, tanh, ReLU).
- $m$: Number of input points $x_j$.
- $\xi_{ij}^k$: Weight for the $j$-th input in the $i$-th branch of the $k$-th trunk.
- $u(x_j)$: Value of the input function $u$ at point $x_j$.
- $\theta_i^k$: Bias for the $i$-th branch in the $k$-th trunk.
- $w_k$: Weight vector for the $k$-th trunk.
- $y$: Output variable.
- $\zeta_k$: Bias for the $k$-th trunk.
- $\epsilon$: Desired approximation accuracy.

- $u$: This represents the input function or input signal. It is the function that the operator $G$ acts upon. In the context of the DeepONet architecture, the network does not take the entire function u as input directly. Instead, it uses the values of the input function u evaluated at a fixed number of specific locations called "sensors" {x1, x2, ..., xm}.
These sensor locations are the same for all input functions u in the training dataset.
The input function u is considered to belong to a set $V$, which is a compact set in the space of continuous functions $C(K1)$, where $K1$ is a compact set in a Banach space X.

- $G$: This symbol represents a nonlinear continuous operator. An operator is a mathematical mapping that takes an input function and transforms it into an output function. 

- $G(u)$: This notation refers to the output function that results from applying the operator G to the input function $u$ . In the examples discussed in the paper, $G(u)$ is the solution function $(s(x)$ or $s(x, t))$ of a differential equation (like an ODE or PDE) when $u$ is the input or source term for that equation. The operator $G$ maps a compact set $V$ of input functions (in $C(K1)$) to a set of output functions (in $C(K2)$, where $K2$ is a compact set in Rd). 

- $y$: This represents a specific location or point in the domain of the output function $G(u)$. It is the location where the value of the output function $G(u)$ is desired. 

- $G(u)(y)$: This is the real number value that represents the output function $G(u)$ evaluated at the specific location $y$. This is the target output that the DeepONet network is trained to predict, given the sensed input $u(x_i)$ and the location $y$.

### Why DeepONet is Needed

1. **The Universal Approximation Theorem for Operators has Limitations in Practice:**  
   While it is known that neural networks can accurately approximate any nonlinear continuous operator, the Universal Approximation Theorem (specifically Theorem 1) only guarantees a small approximation error for a sufficiently large network. It does not consider the important optimization error and generalization error.

2. **Optimization and Generalization Errors are Crucial:**  
   These other error types (optimization and generalization errors) are equally important and often dominant contributions to the total error in practice. The theorem provides a theoretical basis but does not inform how to learn operators effectively. Networks that are practically useful must be easy to train (small optimization error) and generalize well to unseen data (small generalization error).

3. **Generic Architectures are Insufficient for Operators:**  
   Similar to how generic fully-connected neural networks (FNNs) perform poorly compared to specialized architectures like Convolutional Neural Networks (CNNs) for tasks like image classification, despite FNNs being universal function approximators, a specific architecture is needed for effectively learning operators. Directly using classical networks like FNNs by concatenating the inputs (sensed function values and output location) doesn't leverage any specific structure of the problem.

4. **Need for a Specific Architecture to Reduce Generalization Error:**  
   To realize the universal approximation theorem for operators in practice, and to learn operators accurately and efficiently from data, a specific network architecture is required that minimizes the total error, particularly by achieving smaller generalization error.

Therefore, **DeepONet** is proposed as a specific network architecture designed to address these practical challenges. It aims to learn operators accurately and efficiently from a relatively small dataset. Its design, consisting of two sub-networks – a branch net for the input function values at sensors and a trunk net for the output locations – is intended to significantly reduce the generalization error compared to generic fully-connected networks. This specific architecture provides an inductive bias consistent with the problem structure (the output \( G(u)(y) \) depends independently on the input function \( u \) and the output location \( y \)).

### Define RBF Kernel
- **Purpose**: Implements the RBF kernel for GRF covariance.
- **Functionality**:
  - Computes  $ ( k(x_1, x_2) = \exp\left(-\frac{\|x_1 - x_2\|^2}{2l^2}\right) ) $ with `length_scale=0.2`.
  - `partial_rbf_kernel` simplifies calls by fixing the length scale.
- **JAX-Specific**: Uses `jnp` for array operations and broadcasting (`x1[:, None, :] - x2[None, :, :]`) for efficient pairwise distance computation, enabling autodiff and JIT compatibility.

The Radial Basis Function (RBF) kernel, also known as the Gaussian kernel, is a widely used kernel function in machine learning and statistical modeling, particularly for Gaussian Random Fields (GRFs) and kernel-based methods like Gaussian Processes (GPs). Its primary purpose is to define a covariance structure between points in a continuous domain, capturing smooth and stationary correlations. The RBF kernel is implemented to model the covariance function of a GRF, which is essential for generating synthetic data representing antiderivative operators in the DeepONet framework.
Mathematical Definition
The RBF kernel is defined as:
$$k(x_1, x_2) = \exp\left(-\frac{\|x_1 - x_2\|^2}{2l^2}\right)$$
where:

$x_1$ and $x_2$ are two points in the input space (e.g., coordinates in $[0, 1]$).
$\|x_1 - x_2\|$ is the Euclidean distance between $x_1$ and $x_2$.
$l$ is the length scale parameter, which controls the rate of correlation decay. A smaller $l$ implies a shorter range of correlation (faster decay), while a larger $l$ allows for longer-range dependencies.

In our implementation, the length scale is fixed at $l = 0.2$, which suggests a moderate correlation length suitable for the spatial domain of the problem (e.g., $[0, 1]$).
Key Properties

Smoothness: The exponential form ensures the kernel is infinitely differentiable, leading to smooth sample paths of the GRF. This is critical for modeling functions like antiderivatives, which are inherently smooth. 

Stationarity: The RBF kernel depends only on the distance $\|x_1 - x_2\|$, making it a stationary kernel. This implies that the covariance structure is translation-invariant, which is a reasonable assumption for many physical or synthetic systems. 

Positive Definiteness: The kernel generates a positive semi-definite covariance matrix, a requirement for valid GRF sampling via the Cholesky decomposition or Karhunen-Loève expansion. 

Normalization: At zero distance ($x_1 = x_2$), $k(x_1, x_1) = 1$, providing a natural normalization where the variance at any point is 1 (assuming a unit variance GRF).

## Implementation Details


Broadcasting: The expression x1[:, None, :] - x2[None, :, :] computes pairwise differences between all points in $x1$ and $x2$ using NumPy-style broadcasting. This avoids explicit loops, making the code vectorized and scalable to large datasets. 

JAX Compatibility: By using jnp (JAX's NumPy), the kernel supports automatic differentiation (e.g., for gradient-based optimization) and just-in-time (JIT) compilation, enhancing performance on accelerators like GPUs or TPUs. 

Partial Function: The partial_rbf_kernel function fixes the length scale at 0.2, simplifying calls and ensuring consistency across the notebook. This is implemented using functools.partial, a Python utility that pre-sets arguments of a function.

```python
def rbf_kernel(x1: jnp.ndarray, x2: jnp.ndarray, length_scale: float = grf_length_scale) -> jnp.ndarray:
  """
  Computes the RBF kernel between two sets of points.

  This is a flexible, merged function. If you don't provide a `length_scale`,
  it will automatically use the default value.

  Args:
    x1: First set of points, shape (N, D).
    x2: Second set of points, shape (M, D).
    length_scale: The length scale of the kernel. Defaults to DEFAULT_GRF_LENGTH_SCALE.

  Returns:
    An (N, M) kernel matrix.
  """
  # x1 = jnp.atleast_2d(x1)
  # x2 = jnp.atleast_2d(x2)
  diff = x1[:, None, :] - x2[None, :, :]
  sq_dist = jnp.sum(diff**2, axis=-1)
  exp_value =  jnp.exp(-0.5 * sq_dist / length_scale**2)

  return exp_value



def partial_rbf_kernel(x1: jnp.ndarray, x2: jnp.ndarray) -> jnp.ndarray:
  """
  This is a wrapper function for our convenience.
  It calls the main rbf_kernel but automatically provides the
  fixed length_scale we defined earlier.
  """
  # We call the original function and pass along the x1 and x2 we received,
  # plus the pre-set grf_length_scale.
  return rbf_kernel(x1, x2, length_scale=grf_length_scale)
```

### Sample from Gaussian Random Field (GRF)
- **Purpose**: Generates random, smooth functions \( u(x) \) for DeepONet input.
- **Functionality**:
  - Computes Gram matrix $ K $ using the RBF kernel.
  - Applies Cholesky decomposition $( K = L L^T )$ and transforms normal samples $ z \sim \mathcal{N}(0, I) $ to GRF samples $ u = L z$.
  - Adds jitter (`1e-3`) for numerical stability.
- **JAX-Specific**: Uses `jax.random.normal` for random sampling, `jnp.linalg.cholesky` for decomposition, and `jnp` operations for efficient, differentiable computations.

The Gaussian Random Field (GRF) sampling process is a key component of the notebook, designed to generate random, smooth functions $ u(x) $ that serve as inputs to the DeepONet model. These functions represent realizations of a stochastic process with a specified covariance structure, making them ideal for training operator networks on diverse, continuous data. The GRF is particularly suited for modeling physical systems where smoothness and spatial correlation are important, such as in the antiderivative problem that we are addressing.
Mathematical Foundation
A GRF is a stochastic process where every finite collection of points follows a multivariate Gaussian distribution. For a set of points $ x_1, x_2, \ldots, x_n $ in the domain (e.g., $[0, 1]$), the GRF $ u(x) $ is defined by:

A mean function (often zero for simplicity: $ \mu(x) = 0 $).
A covariance function $ k(x_i, x_j) $, which in your case is the RBF kernel:
$$k(x_i, x_j) = \exp\left(-\frac{\|x_i - x_j\|^2}{2l^2}\right)$$
with a fixed length scale $ l = 0.2 $.

The covariance matrix $ K $, where $ K_{ij} = k(x_i, x_j) $, fully specifies the joint distribution. Samples from the GRF are drawn using the property that if $ z \sim \mathcal{N}(0, 1) $ is a vector of independent standard normal variables, then $ u = L z $ follows $ \mathcal{N}(0, K) $, where $ L $ is the lower triangular Cholesky factor of $ K $ (i.e., $ K = L L^T $).

## Implementation Details


Covariance Matrix Construction: The kernel matrix $ K $ is computed using the partial_rbf_kernel defined earlier, evaluated over a grid of points $ x $.

Cholesky Decomposition: The jnp.linalg.cholesky function computes $ L $ from $ K $. To ensure numerical stability, a small jitter (e.g., $ 1e-3 \cdot I $) is added to the diagonal of $ K $ before decomposition, preventing issues with near-singular matrices.

Random Sampling: jax.random.normal generates $ z $, a vector of i.i.d. standard normal samples, which is then transformed by $ L $ to produce the GRF sample $ u $.

JAX-Specific Features: The use of jnp ensures compatibility with JAX’s autodiff and JIT compilation, enabling efficient computation and potential acceleration on GPUs or TPUs. The vectorized operations also handle batch sampling seamlessly.

```python
def sample_grf(key: jax.random.key, kernel_fn: Callable, eval_pts: jnp.ndarray, num_samples: int, jitter: float = 1e-3) -> jnp.ndarray:
  """
  Samples functions from a zero-mean Gaussian Random Field (GRF).
  This creates our smooth, random-looking input functions u(x).
  """
  # --- Step 1: Define the "similarity" between all points ---
  # The Gram matrix K defines the covariance. A high value K[i, j] means
  # points x[i] and x[j] should have similar values in our random function.
  K = kernel_fn(eval_pts, eval_pts)

  # --- Step 2: Make the matrix numerically stable ---
  # Cholesky decomposition requires the matrix to be strictly positive definite.
  # Adding a tiny "jitter" to the diagonal ensures this.
  K_stable = K + jitter * jnp.eye(K.shape[0])

  # --- Step 3: Get the "matrix square root" ---
  # The Cholesky decomposition gives us a matrix L where L @ L.T = K_stable.
  # This L is the key to transforming simple noise into structured, smooth functions.
  L = jnp.linalg.cholesky(K_stable)

  # --- Step 4: Generate simple, uncorrelated random noise ---
  # These are just random numbers from a standard normal distribution (mean=0, std=1).
  # Shape is (number_of_points, number_of_functions_to_create).
  z = jax.random.normal(key, shape=(K.shape[1], num_samples))

  # --- Step 5: Create the smooth functions ---
  # This is the magic step. By multiplying the simple noise `z` by our
  # transformation matrix `L`, we create correlated samples that respect
  # the smoothness defined by our kernel.
  # We transpose (.T) at the end so the final shape is (num_samples, num_points).
  samples = jnp.dot(L, z).T
  return samples
```

### Numerical Integration for Antiderivative
- **Purpose**: Computes the antiderivative $ G(u)(y) = \int_0^y u(t) \, dt $ as the target output for DeepONet training.
- **Functionality**:
  - Solves the ODE $ \frac{dG}{dy} = u(y) $, $ G(0) = 0 $, using `odeint`.
  - Interpolates $ u(t) $ at arbitrary points using `jnp.interp` for ODE evaluation.
- **JAX-Specific**: Uses `jax.experimental.ode.odeint` for numerical integration, `jnp.interp` for interpolation, and functional programming (lambda functions) for compatibility with JAX’s autodiff.

```python
def compute_antiderivative(u_function: jnp.ndarray, x_fine_grid: jnp.ndarray, y_locations: jnp.ndarray) -> jnp.ndarray:
  """
  Computes the antiderivative of a single function u at multiple y locations.
  This solves the ODE: dG/dy = u(y), with the initial condition G(0) = 0.
  """
  # Ensure y_locations is a 1D array for the solver.
  y_locations = jnp.atleast_1d(y_locations).flatten()

  # The ODE solver needs a function that can evaluate u(t) at ANY time t.
  # jnp.interp creates this function for us from our discrete data points.
  # The ODE solver will call this many times.
  # The `_` is a placeholder for the state `x`, which we don't need here.
  u_as_a_callable_function = lambda t, _: jnp.interp(t, x_fine_grid.flatten(), u_function)

  # The solver needs to know all the points where we want the answer.
  # Since our integral starts at 0, we must include 0.0 as the first point.
  evaluation_points = jnp.concatenate([jnp.array([0.0]), y_locations])

  # Call the ODE solver.
  # It solves dG/dy = u(y) starting with G=0.0 at time t=0.0.
  # It gives us the solution G at all the `evaluation_points`.
  solution = odeint(u_as_a_callable_function, 0.0, evaluation_points)

  # The first value in `solution` corresponds to t=0.0, which we don't need.
  # We return all the other values, which correspond to our `y_locations`.
  return solution[1:]
```

## Explanation of Numerical Integration for Antiderivative

The numerical integration for antiderivative section of the notebook is designed to compute the antiderivative $ G(u)(y) = \int_0^y u(t) \, dt $ as the target output for training the DeepONet model. This process transforms the input function $ u(t) $, sampled from a Gaussian Random Field (GRF), into its integral over the interval $[0, y]$, effectively modeling an integral operator. This is a critical step, as the DeepONet learns to approximate this operator mapping from $ u(t) $ to $ G(u)(y) $, aligning with the goal of operator learning for differential equations or physical systems.

Mathematical Foundation
The antiderivative $ G(u)(y) $ represents the definite integral of $ u(t) $ from 0 to $ y $, satisfying the initial value problem:
$$\frac{dG}{dy} = u(y), \quad G(0) = 0$$

$ u(y) $ is the input function, typically a smooth realization from the GRF.
The solution $ G(y) $ accumulates the area under $ u(t) $ up to $ y $, making it a natural target for an operator network like DeepONet, which maps functions to functions.

Since $ u(t) $ is defined on a discrete grid (e.g., $ x_{\text{fine}} $), numerical methods are required to evaluate the integral at arbitrary points $ y $. This involves interpolation to estimate $ u(y) $ between grid points and integration to compute $ G(y) $.

## Implementation Details

ODE Integration: The jax.experimental.ode.odeint function solves the differential equation numerically. It treats $ \frac{dG}{dy} = u(y) $ as an ODE, stepping from $ y = 0 $ to the desired $ y $ values with $ G(0) = 0 $.

Interpolation: The jnp.interp function interpolates $ u(t) $ from the fine grid $ x_{\text{fine}} $ to the evaluation points $ y $, ensuring $ u(y) $ is available for the ODE solver. Linear interpolation is typically used for simplicity and efficiency.

JAX-Specific Features: The use of jnp arrays and functional programming (e.g., lambda functions) ensures compatibility with JAX’s automatic differentiation and just-in-time (JIT) compilation. This allows the integration process to be differentiable, which is useful for end-to-end training of the DeepONet.

### Vectorized Operations
- **Purpose**: Vectorizes interpolation and antiderivative computations for efficiency.
- **Functionality**:
  - `vmap_interp`: Interpolates multiple \( u(x) \) from `x_fine_grid` to `x_sensors`.
  - `vmap_compute_antiderivative_single_u`: Computes \( G(u)(y) \) for one \( u \) at multiple \( y \).
  - `vmap_compute_antiderivative_batch_u_y`: Computes \( G(u)(y) \) for multiple \( u \) and \( y \).
- **JAX-Specific**: Uses `jax.vmap` for parallel computation across arrays, with `in_axes` and `out_axes` to control vectorization, leveraging GPU/TPU acceleration.

```python
#  For creating branch inputs
# We want to interpolate many different u_functions (the 3rd argument), but always
# using the same x_sensors (1st arg) and x_fine_grid (2nd arg).
# So, in_axes should be (None, None, 0).
vmap_interp = jax.vmap(
    jnp.interp,
    in_axes=(None, None, 0),
    out_axes=0
)

# For calculating the final answers
# This is a double-vmap. First, we create a worker that can handle one `u`
# and a batch of `y` locations.
vmap_compute_antiderivative_single_u = jax.vmap(
    compute_antiderivative,
    in_axes=(None, None, 0), # Loop over y_locations (3rd arg)
    out_axes=0
)

# Now, we wrap that worker in *another* vmap to handle a batch of `u` functions
# and a batch of `y_locs_batch`.
vmap_compute_antiderivative_batch_u_y = jax.vmap(
    vmap_compute_antiderivative_single_u,
    in_axes=(0, None, 0), # Loop over u_functions (1st arg) and y_locs_batch (3rd arg)
    out_axes=0
)
```

### Main Data Generation Function
- **Purpose**: Generates DeepONet training/test data: branch inputs (\( u(x) \) at sensor points), trunk inputs (\( y \)), and outputs (\( G(u)(y) \)).
- **Functionality**:
  - Samples \( u(x) \) from GRF on `x_fine_grid`.
  - Interpolates to `x_sensors` for branch inputs.
  - Generates random \( y \)-locations and computes \( G(u)(y) \).
  - Formats data for DeepONet training.
- **JAX-Specific**: Uses `jax.random.split` for reproducible randomness, `jnp.repeat` and `jnp.reshape` for array formatting, and `vmap` for efficient vectorized computations.

```python
def generate_data_antiderivative(key: jax.random.key, num_u_samples: int, num_y_per_u: int,
                                 x_fine_grid: jnp.ndarray, x_sensors: jnp.ndarray, grf_kernel_fn: Callable,
                                 domain_start_val: float, domain_end_val: float
                                 ) -> Tuple[jnp.ndarray, jnp.ndarray, jnp.ndarray, jnp.ndarray]:
  """
  Generates data for DeepONet training using vmap for high performance.
  """
  # Split the random key for reproducible randomness in different steps.
  key_grf, key_y_locs = jax.random.split(key)

  #  1: Create the Input Functions, u(x)
  u_samples_fine = sample_grf(key_grf, grf_kernel_fn, x_fine_grid, num_u_samples)

  # 2: Create the Branch Net Inputs (using vmap_interp)
  # This single line replaces the entire first for-loop.
  branch_inputs_raw = vmap_interp(x_sensors.flatten(), x_fine_grid.flatten(), u_samples_fine)

  # 3: Create the Trunk Net Inputs
  y_locs_batch = jax.random.uniform(key_y_locs,
                                    shape=(num_u_samples, num_y_per_u),
                                    minval=domain_start_val,
                                    maxval=domain_end_val)

  #  4: Calculate the "Correct Answers" (using double-vmap )
  # This single line replaces the entire second for-loop.
  outputs_raw = vmap_compute_antiderivative_batch_u_y(u_samples_fine, x_fine_grid, y_locs_batch)

  # 5: Format the Data for DeepONet Training
  branch_inputs_final = jnp.repeat(branch_inputs_raw, num_y_per_u, axis=0)
  trunk_inputs_final = y_locs_batch.reshape(-1, 1)
  outputs_final = outputs_raw.reshape(-1, 1)

  print("Data generation complete")
  return branch_inputs_final, trunk_inputs_final, outputs_final, u_samples_fine
```

### Define DeepONet Models
- **Purpose**: Defines DeepONet architectures to learn the antiderivative operator.
- **Functionality**:
  - `DeepONetUnstacked`: Single branch-trunk pair with dot product and bias.
  - `DeepONetStacked`: Multiple branch-trunk pairs (5 stacks) summed together.
- **JAX-Specific**: Uses `flax.linen` for modular neural networks, `nn.Dense` and `nn.relu` for layers, and `jnp.sum`/`jnp.stack` for combining outputs.

The DeepONet and BaselineFNN models defined in your notebook are neural network architectures tailored for operator learning, specifically designed to approximate the antiderivative operator $ G(u)(y) = \int_0^y u(t) \, dt $. These models leverage the Flax library within the JAX ecosystem, enabling efficient computation, automatic differentiation, and potential acceleration on hardware like GPUs or TPUs. Below, we explore the DeepONetUnstacked, DeepONetStacked, and BaselineFNN classes in detail, highlighting their purpose, functionality, and implementation specifics.
DeepONetUnstacked

Purpose: Implements a single-branch DeepONet, a neural operator architecture that learns mappings between infinite-dimensional function spaces. It approximates $ G(u)(y) $ by processing sensor measurements of $ u(t) $ (branch input) and evaluation points $ y $ (trunk input).
Functionality:

Branch Network: Takes $ u(x_i) $ at sensor locations $ x_i $, applying a series of dense layers with ReLU activation to extract features. The architecture is defined by branch_features, a tuple specifying the number of neurons per layer (e.g., (40, 40, 40)).
Trunk Network: Processes $ y $, the output coordinate, through a parallel set of dense layers with ReLU activation, guided by trunk_features.
Combination: The branch and trunk outputs are combined via a dot product, followed by a learnable bias term, to produce the scalar output $ G(u)(y) $.

DeepONetStacked

Purpose: Extends the DeepONetUnstacked model by stacking multiple DeepONet branches, increasing the model’s capacity to capture complex operator behaviors. This is particularly useful for non-linear or high-dimensional operators.
Functionality:

Multiple Stacks: Iterates num_stacks (default 5) times, each stack processing the branch and trunk inputs independently with named layers (e.g., stack_0_branch_dense_0) to avoid parameter conflicts.
Branch and Trunk Networks: Each stack applies the same architecture as DeepONetUnstacked, with separate weights for each layer.
Combination: Outputs from all stacks are summed after individual dot product and bias operations, aggregating the contributions to predict $ G(u)(y) $.

@nn.compact is to simplify how we build neural networks. It lets us define the entire architecture—layers, activations, and outputs—right inside the __call__ method of a Flax module. This handles both initializing parameters like weights and biases and running the forward pass, all in one place.

BaselineFNN

Purpose: Provides a baseline feedforward neural network (FNN) to compare against DeepONet, learning the same operator by concatenating sensor inputs and $ y $ into a single vector.
Functionality:

Input Concatenation: Combines $ u(x_i) $ at sensor points and $ y $ into a single input vector.
Hidden Layers: Applies depth layers, each with width neurons and ReLU activation, to process the concatenated input.
Output: A final dense layer with 1 neuron produces a scalar $ G(u)(y) $, with squeeze() removing singleton dimensions for loss computation.

```python
# Define unstacked DeepONet
class DeepONetUnstacked(nn.Module):
  branch_features: tuple
  trunk_features: tuple

  @nn.compact
  def __call__(self, branch_input, trunk_input):
    # Branch network: processes u(x_i)
    x = branch_input
    for f in self.branch_features[:-1]: # all but last layer
        x = nn.Dense(f)(x)
        x = nn.relu(x)
    branch_out = nn.Dense(self.branch_features[-1])(x) # for last layer

    # Trunk network: processes y
    y = trunk_input
    for f in self.trunk_features[:-1]:
        y = nn.Dense(f)(y)
        y = nn.relu(y)
    trunk_out = nn.Dense(self.trunk_features[-1])(y)

    # Combine with dot product and bias
    output = jnp.sum(branch_out * trunk_out, axis=-1, keepdims=True)
    bias = self.param('bias', nn.initializers.zeros, (1,))
    # bias = 0
    #bias = self.param('bias', nn.initializers.zeros, (branch_input.shape[0],1))
    #print(bias.shape)

    return output + bias

  # Stacked DeepONet
  # Stacked DeepONet
class DeepONetStacked(nn.Module):
  branch_features: tuple
  trunk_features: tuple
  num_stacks: int = 5 # Number of DeepONet branches

  @nn.compact
  def __call__(self, branch_input, trunk_input):
    outputs = []
    for i in range(self.num_stacks):
      # Branch network for stack i
      x = branch_input
      for j, f in enumerate(self.branch_features[:-1]):
          x = nn.Dense(f, name=f'stack_{i}_branch_dense_{j}')(x)
          x = nn.relu(x)
      branch_out = nn.Dense(self.branch_features[-1], name=f'stack_{i}_branch_out')(x)

      # Trunk network for stack i
      y = trunk_input
      for j, f in enumerate(self.trunk_features[:-1]):
          y = nn.Dense(f, name=f'stack_{i}_trunk_dense_{j}')(y)
          y = nn.relu(y)
      trunk_out = nn.Dense(self.trunk_features[-1], name=f'stack_{i}_trunk_out')(y)

      # Combine with dot product and bias
      output = jnp.sum(branch_out * trunk_out, axis=-1, keepdims=True)
      bias = self.param(f'stack_{i}_bias', nn.initializers.zeros, (1,))
      outputs.append(output + bias)

    # Sum outputs from all stacks
    return jnp.sum(jnp.stack(outputs, axis=0), axis=0)



    # --- The Generalized FNN Model Definition ---
class BaselineFNN(nn.Module):
  depth: int  # The number of hidden layers
  width: int  # The number of neurons in each hidden layer

  @nn.compact
  def __call__(self, x):
      # x is the concatenated input: [u_sensors, y_loc]

      # Hidden layers
      for _ in range(self.depth):
          x = nn.Dense(features=self.width)(x)
          x = nn.relu(x)

      # Final output layer to produce a single scalar value
      x = nn.Dense(features=1)(x)
      return x.squeeze() # Squeeze to get shape (batch,) for easier loss calculation
```

### Loss Function
- **Purpose**: Defines the mean squared error (MSE) loss for training.
- **Functionality**: Computes $\text{MSE} = \text{mean}((pred - true)^2) $.
- **JAX-Specific**: Uses `jnp.mean` for differentiable loss computation, adhering to JAX’s functional programming.

```python
# Loss function
def mse_loss(pred, true):
  return jnp.mean((pred - true) ** 2)
```

### Training and Evaluation Functions
- **Purpose**: Defines functions for training and evaluating the DeepONet model.
- **Functionality**:
  - `train_step`: Computes loss and gradients, updates parameters for one batch.
  - `eval_model`: Evaluates MSE loss on test data.
- **JAX-Specific**: Uses `jax.jit` for compilation, `jax.value_and_grad` for gradient computation, and `static_argnums` to handle Flax models.

The training and evaluation functions section of your notebook provides the core mechanisms for optimizing and assessing the DeepONet model, specifically tailored to learn the antiderivative operator $ G(u)(y) = \int_0^y u(t) \, dt $. These functions leverage JAX’s advanced features to ensure efficient computation, automatic differentiation, and scalability, making them integral to the training pipeline of the DeepONetUnstacked and DeepONetStacked models.

Training Function: train_step

Purpose: Updates the model parameters for a single batch of data by computing the loss and applying gradients using an optimizer.
Functionality:

Takes a TrainState object (containing model parameters and optimizer state) and a batch of data (branch_batch, trunk_batch, output_batch).
Defines a loss function based on mean squared error (MSE) between the model’s prediction and the true antiderivative output.
Uses jax.value_and_grad to compute both the loss value and its gradients with respect to the parameters.
Applies the gradients to update the TrainState using the optimizer (e.g., Adam).


Evaluation Function: eval_model

Purpose: Computes the mean squared error (MSE) loss on a test dataset to assess model performance.
Functionality:

Takes model parameters, the Flax model instance, and a tuple of test data (test_branch, test_trunk, test_output).
Applies the model to predict outputs for the test inputs and calculates the MSE loss against the true outputs.
Returns a scalar loss value for the entire dataset.


JAX-Specific: Also uses jax.jit for compilation, with static_argnums=(1,) to treat the model as static. This ensures the function is optimized for repeated evaluations without recompiling the model structure, enhancing efficiency during validation or testing.

Why Use JIT?: Similar to train_step, JIT compilation in eval_model optimizes the evaluation process by pre-computing the computation graph. This is particularly beneficial when evaluating on large test sets or performing multiple validation checks, as it reduces latency and leverages hardware acceleration, making the training loop more responsive.

```python
@jax.jit
def train_step(state, batch):
  """Trains the model for one step."""
  branch_input, trunk_input, target = batch
  def loss_fn(params):
    pred = state.apply_fn({'params': params}, branch_input, trunk_input)
    return mse_loss( pred.reshape(-1,1),  target)

  # grad_fn = jax.grad(loss_fn)
  # grads = grad_fn(state.params)
  loss, grads = jax.value_and_grad(loss_fn)(state.params)
  state = state.apply_gradients(grads=grads)
  # loss = loss_fn(state.params) # Calculate loss again for logging
  return state, loss

def eval_model(params, model, data):
  """Evaluates the model on a dataset."""
  branch_input, trunk_input, target = data
  predictions = model.apply({'params': params}, branch_input, trunk_input)
  return mse_loss( predictions.flatten(), target.flatten()) # Ensure shapes match

jitted_eval_model = jax.jit(eval_model, static_argnums=(1,))
```

```python
# Create FNN-Specific Training and Evaluation Functions ---

@jax.jit
def train_step_fnn(state, batch):
  """A dedicated training step for the BaselineFNN."""
  branch_input, trunk_input, target = batch

  # Concatenate inputs for the FNN model
  fnn_input = jnp.concatenate([branch_input, trunk_input], axis=1)

  def loss_fn(params):
    # Note: The FNN model takes a single input tensor
    pred = state.apply_fn({'params': params}, fnn_input)
    # The FNN model already squeezes the output, so we reshape the target to match
    return mse_loss(pred, target.flatten())

  loss, grads = jax.value_and_grad(loss_fn)(state.params)
  state = state.apply_gradients(grads=grads)
  return state, loss

def eval_model_fnn(params, model, data):
  """A dedicated evaluation function for the BaselineFNN."""
  branch_input, trunk_input, target = data

  # Concatenate inputs for the FNN model
  fnn_input = jnp.concatenate([branch_input, trunk_input], axis=1)

  predictions = model.apply({'params': params}, fnn_input)
  return mse_loss(predictions, target.flatten())

jitted_eval_model_fnn = jax.jit(eval_model_fnn, static_argnums=(1,))
```

### Training Loop
- **Purpose**: Trains the DeepONet model and evaluates performance.
- **Functionality**:
  - Trains for 10,000 epochs with batch size 256.
  - Shuffles data each epoch and computes train/test MSE losses.
  - Logs losses every 100 epochs (e.g., Epoch 100: Train Loss: 0.000045, Test Loss: 0.000158).
- **JAX-Specific**: Uses `jax.random.permutation` for shuffling, `jax.jit` for fast training/evaluation, and batching for efficient computation.

```python

unstacked_train_losses = []
unstacked_test_losses = []
stacked_train_losses = []
stacked_test_losses = []
base_train_losses = []
base_test_losses = []


num_train_samples = train_branch_data.shape[0]
num_batches = num_train_samples // batch_size
```

```python
print("Starting training for all three models...")
for epoch in range(num_epochs):
  # Shuffle data once per epoch for all models
  key_shuffle, _ = jax.random.split(jax.random.PRNGKey(epoch))
  perm = jax.random.permutation(key_shuffle, train_branch_data.shape[0])
  branch_shuffled, trunk_shuffled, output_shuffled = train_branch_data[perm], train_trunk_data[perm], train_output_data[perm]

  # Initialize epoch losses
  unstacked_epoch_loss = 0.0
  stacked_epoch_loss = 0.0
  base_epoch_loss = 0.0

  for i in range(num_batches):
    start_idx, end_idx = i * batch_size, (i + 1) * batch_size
    batch = (
        branch_shuffled[start_idx:end_idx],
        trunk_shuffled[start_idx:end_idx],
        output_shuffled[start_idx:end_idx]
    )

    # Train unstacked model
    unstacked_state, loss = train_step(unstacked_state, batch)
    unstacked_epoch_loss += loss

    # Train stacked model
    stacked_state, loss = train_step(stacked_state, batch)
    stacked_epoch_loss += loss

    # Train baseline FNN model using its dedicated train_step
    base_state, loss = train_step_fnn(base_state, batch)
    base_epoch_loss += loss

  # --- Log training losses for the epoch ---
  unstacked_train_losses.append(unstacked_epoch_loss / num_batches)
  stacked_train_losses.append(stacked_epoch_loss / num_batches)
  base_train_losses.append(base_epoch_loss / num_batches)

  # --- Compute and log test losses for the epoch ---
  # DeepONets use the original eval function
  unstacked_test_loss = jitted_eval_model(unstacked_state.params, unstacked_model, (test_branch_data, test_trunk_data, test_output_data))
  stacked_test_loss = jitted_eval_model(stacked_state.params, stacked_model, (test_branch_data, test_trunk_data, test_output_data))
  unstacked_test_losses.append(unstacked_test_loss)
  stacked_test_losses.append(stacked_test_loss)

  # FNN uses its dedicated eval function
  base_test_loss = jitted_eval_model_fnn(base_state.params, base_model, (test_branch_data, test_trunk_data, test_output_data))
  base_test_losses.append(base_test_loss)

  if (epoch + 1) % 100 == 0:
    print(f"--- Epoch {epoch + 1} ---")
    print(f"Unstacked: Train Loss: {unstacked_train_losses[-1]:.6f}, Test Loss: {unstacked_test_losses[-1]:.6f}")
    print(f"Stacked:   Train Loss: {stacked_train_losses[-1]:.6f}, Test Loss: {stacked_test_losses[-1]:.6f}")
    print(f"FNN:       Train Loss: {base_train_losses[-1]:.6f}, Test Loss: {base_test_losses[-1]:.6f}")
```


# Validation of Trained DeepONet Model

This section serves as a critical validation step for our trained DeepONet model. After training the model on complex functions sampled from a Gaussian Random Field (GRF), we now test its ability to generalize to simpler, well-defined functions that it has not seen before.

## Purpose
The goal is to verify that the model has learned the underlying mathematical operator of integration (i.e., $ G(u) = \int u(x) \, dx $), rather than just memorizing the statistical patterns of the training data. By feeding it simple linear functions, for which we know the exact analytical antiderivative, we can precisely measure the model's accuracy and generalization performance.

## Plotted Values and Their Computation
Each of the plots generated below visualizes three different curves for a given linear input function $ u(x) $. Here is a breakdown of what each curve represents:

### True (Analytical) - Dotted Black Line
- **What it is**: This is the absolute ground truth. It is the mathematically perfect antiderivative of the input function $ u(x) $.
- **How it's computed**: This solution is derived by hand using calculus. For a general linear function $ u(x) = ax + b $, its antiderivative starting from zero is:
  $$
  G(u)(y) = \int_0^y (ax + b) \, dx = \frac{a}{2} y^2 + b y
  $$
  In the code, these analytical solutions are pre-defined in the `true_antiderivatives` list of lambda functions.

### True (Numerical) - Solid Blue Line
- **What it is**: This is the antiderivative of $ u(x) $ as calculated by our own numerical integration function (`vmap_compute_antiderivative_single_u`).
- **How it's computed**: It uses numerical methods (based on `jnp.cumsum` and `jnp.interp`) to approximate the integral of the input function $ u(x) $.
- **Its Purpose**: This curve serves as a sanity check for our data generation pipeline. If this solid blue line perfectly overlaps the dotted black analytical line, it confirms that our numerical integration code is accurate.

### Predicted - Dashed Red Line
- **What it is**: This is the output of our trained DeepONet model. This is the curve we are evaluating.
- **How it's computed**:
  - The input function $ u(x) $ is evaluated at the predefined `x_sensors` locations. This vector of sensor readings becomes the input to the Branch Net.
  - The `y_vis` locations (the points where we want to know the answer) are fed as input to the Trunk Net.
  - These two inputs are processed by the trained model (`selected_model.apply(...)`) to produce the final predicted values for the antiderivative.



```python
# Define linear functions u(x) = ax + b
linear_functions = [
    lambda x: 2 * x + 1,  # u_1(x) = 2x + 1
    lambda x: -x + 2,    # u_2(x) = -x + 2
    lambda x: x          # u_3(x) = x
]
labels = ['u(x) = 2x + 1', 'u(x) = -x + 2', 'u(x) = x']

# Compute true antiderivatives analytically: G(u)(y) = a/2 * y^2 + b*y
true_antiderivatives = [
    lambda y: y**2 + y,         # G(u_1)(y) = y^2 + y
    lambda y: -0.5 * y**2 + 2*y, # G(u_2)(y) = -0.5*y^2 + 2y
    lambda y: 0.5 * y**2        # G(u_3)(y) = 0.5*y^2
]

# Evaluation points
y_vis = jnp.linspace(domain_start, domain_end, 50)



# --- Loop to create a separate figure for each function ---
for i, (u_fn, true_G, label) in enumerate(zip(linear_functions, true_antiderivatives, labels)):

    # --- 1. Create a new, single-panel figure for this test case ---
    plt.figure(figsize=(10, 6))

    # --- 2. Perform calculations for this function ---
    # Evaluate u(x) to get sensor readings for the model input
    u_sample = u_fn(x_fine_grid.flatten())
    u_sensors = u_fn(x_sensors.flatten())

    # Get the true solutions for comparison
    true_s_numerical = vmap_compute_antiderivative_single_u(u_sample, x_fine_grid, y_vis)
    true_s_analytical = true_G(y_vis)

    # Get the model's prediction
    u_sensors_batch = jnp.repeat(u_sensors[None, :], 50, axis=0)
    y_vis_batch = y_vis[:, None]
    pred_s = selected_model.apply({'params': selected_state.params}, u_sensors_batch, y_vis_batch)

    # --- 3. Create the plot for this figure ---
    # Plot the true and predicted antiderivatives G(u)(y)
    plt.plot(y_vis, true_s_analytical, color='black', linestyle=':', linewidth=3, label='True (Analytical)')
    plt.plot(y_vis, true_s_numerical, color='blue', linestyle='-', label='True (Numerical)')
    plt.plot(y_vis, pred_s.flatten(), color='red', linestyle='--', label='Predicted')

    # Configure the plot
    plt.xlabel("y")
    plt.ylabel("G(u)(y)")
    plt.title(f"Antiderivative Comparison for u(x) = {label}\n({model_name} DeepONet)")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # --- 4. Display the completed figure ---
    plt.tight_layout()
    plt.show()
```
